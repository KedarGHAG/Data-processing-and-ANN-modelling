## Extract complete Rainfall data from downloaded zip folder

```python
####  Import necessary libraries
import os
import zipfile
import pandas as pd
import datetime
import re

filename='Station-EDEN_6-R1549451756.zip'

#### regular expression to extract station name and zip number
stnname1 = re.findall('^Station-(.*)-R',filename)
rzipnum1 = re.findall('-R(.*).zip',filename)
stnname = stnname1[0]
rzipnum = rzipnum1[0]
path1='C:\\Data\\Daily Rainfall and Hourly Water Level data'

#### file path from where downloaded zip files need to be taken
zippath_R=path1

#### file path where extracted files need to be stored
extractedzip=path1

#### file path to save data collected from extracted csv files
dR_fpath = 'C:\\Data'
now = datetime.datetime.now()
tstamp=str(now)[11:13]+str(now)[14:16]+str(now)[17:19]+str(now)[-7:]
rzipname = "Station-"+stnname+"-R"+rzipnum+".zip"

#### file path to get extracted files
extractdata = extractedzip +"\eve_data"

#### Read CSV Hourly Water Level [rcsvhWL]
rcsvdR = rzipnum+"_rainfall.csv"

#### Write CSV Daily Water Level [wcsvdWL]
wcsvdR = "dR["+stnname+"]"+tstamp+".csv"

#### Extract data from zipped folder and change directory where rainfall data is downloaded
os.chdir(zippath_R)

#### create file name as per the standard format (E.g. Station-W2-1544019002.zip)
zfr = zipfile.ZipFile(rzipname)
zfr.namelist()

#### file path where extracted files need to be stored
zfr.extractall(extractedzip)

#### Now, read extracted csv file
os.chdir(extractdata)

#### read csv file
df_dR = pd.read_csv(rcsvdR, skiprows = 5, usecols = [0,1], header = None, 
                         names = ['Time', 'Daily rainfall (inches)'],
                         sep = ',', index_col = False)

#### convert daily rainfall from inches to mm and save it into new column
df_dR['Daily rainfall (mm)'] = (df_dR['Daily rainfall (inches)']*25.4).round(2)

#### save values in csv file
os.chdir(dR_fpath)
df_dR.to_csv(wcsvdR,index=False,columns=['Time','Daily rainfall (inches)','Daily rainfall (mm)'])
```

# Extract Rainfall data for Specific period from downloaded zip folder

```python
#### Import necessary libraries
import os
import pandas as pd
import datetime
import re

#### First, define variables and get data from user
getfilepath='C:\\Data'
os.chdir(getfilepath)
getdRcsv='dR[EDEN_6]121203.829826.csv'
firstDate='2016-01-01'
lastDate='2018-01-08'

#### regular expression to extract station name
station=re.findall('^dR\[(.*)]',getdRcsv)
station=station[0]

#### read csv
df_dR=pd.read_csv(getdRcsv,
                  skiprows=1,
                  sep = ',',
                  names=['Time','DailyRainfall(in)','DailyRainfall(mm)'])

#### select all rows in DataFrame between given date range
df_dR=df_dR[(df_dR['Time']>=firstDate)&(df_dR['Time']<=lastDate)]
df_dR=df_dR.reset_index(drop=True)

#### save values in csv file
setfilepath='C:\\Data'
os.chdir(setfilepath)
now=datetime.datetime.now()
timeStamp=str(now)[11:13]+str(now)[14:16]+str(now)[17:19]+str(now)[-7:]
setdRcsv="dR["+station+"]["+firstDate+" to "+lastDate+"]"+timeStamp+".csv"
df_dR.to_csv(setdRcsv,
             index = False,
             columns=['Time','DailyRainfall(in)','DailyRainfall(mm)'])
```
